```python
df = pd.read_csv('Keras - Data.csv')

pd.set_option("display.max_rows", None ,"display.max_columns", None)
```

```python
df1 = pd.DataFrame({'Cement': df.iloc[: , 0], 'Blast Furnace': df.iloc[: , 1],
                   'Fly Ash': df.iloc[: , 2], 'Water': df.iloc[: , 3], 'Superplasticizer': df.iloc[: , 4],
                   'Coarse Aggregate': df.iloc[: , 5], 'Fine Aggregate': df.iloc[: , 6],
                   'Age': df.iloc[: , 7], 'Strength': df.iloc[: , 8]})

df1
```

# Baseline Model (Part A)

```python
X = df1.iloc[: , [0,1,2,3,4,5,6]]

Y = df1.iloc[: , [8]]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .3)
```

```python
model = tf.keras.Sequential()
model.add(layers.Dense(10, activation='relu',
                           input_shape=(X_train.shape[1],)))
model.add(layers.Dense(1))
model.compile(optimizer='adam', loss='mse', metrics=['mse'])
```

```python
epochs = 50

iterations = 50

mse = []

for i in range(iterations):

    history = model.fit(X_train, Y_train,
                    validation_data=(X_test, Y_test),
                    epochs=epochs, verbose=0) 
    mse_history = history.history['mse']
    mse.append(mse_history)
```

```python
mse_history
```

```python
mse_range = np.array(mse_history)

mse_std = mse_range.std()

mse_mean = mse_range.mean()

```

```python
std_mean = pd.DataFrame({'Standard Deviation (MSE)': mse_std, 'Mean (MSE)': mse_mean}, index=[0])

std_mean
```

# Normalized (Part B)

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .3)
```

```python
from sklearn import preprocessing

X_test_norm = preprocessing.normalize(X_test)
X_train_norm = preprocessing.normalize(X_train)
```

```python
model = tf.keras.Sequential()
model.add(layers.Dense(10, activation='relu',
                           input_shape=(X_train_norm.shape[1],)))
model.add(layers.Dense(1))
model.compile(optimizer='adam', loss='mse', metrics=['mse'])
```

```python
epochs = 50

for i in range(50):
    new_history = model.fit(X_train_norm, Y_train,
                validation_data=(X_test_norm, Y_test),
                epochs=epochs, verbose=0) 
    new_mse_history = new_history.history['mse']
    mse.append(new_mse_history)
```

```python
print(new_mse_history)
```

Part B (Mean Squared Error)

```python
new_mse_range = np.array(new_mse_history)

new_mean_mse = new_mse_range.mean()

new_mean_mse
```

Part B(Standard Deviation)

```python
new_mse_std = mse_range.std()

new_mse_std
```

```python
Comparison = pd.DataFrame({'Initial Mean(MSE)': mse_mean, 'After Normalization Mean(MSE)': new_mean_mse},
                         index=['Comparison'])

Comparison
```

The average mean squared error slightly decreased as we normalized the predictors.

# Increase Hidden Layers (Part D)

```python
model = tf.keras.Sequential()
model.add(layers.Dense(10, activation='relu',
                           input_shape=(X_train_norm.shape[1],)))
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1))
model.compile(optimizer='adam', loss='mse', metrics=['mse'])
```

```python
epochs = 100

iterations = 50

inc_HL = []

for i in range(iterations):

    inc_HL_history = model.fit(X_train_norm, Y_train,
                    validation_data=(X_test_norm, Y_test),
                    epochs=epochs, verbose=0) 
    inc_HL_history = inc_HL_history.history['mse']
    inc_HL.append(inc_HL_history)
```

Part D (Mean Squared Error)

```python
inc_HL_range = np.array(inc_HL_history)

inc_HL_mse = inc_HL_range.mean()

inc_HL_mse
```

```python
inc_HL_std = inc_HL_range.std()

inc_HL_std
```
